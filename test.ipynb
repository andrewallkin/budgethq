````python
# Import required libraries
import pytesseract
from PIL import Image
from pdf2image import convert_from_path
import os
from pathlib import Path

# Check if Tesseract is installed
try:
    pytesseract.get_tesseract_version()
    print("✅ Tesseract is installed and working!")
except Exception as e:
    print(f"❌ Tesseract not found: {e}")
    print("Install Tesseract:")
    print("  macOS: brew install tesseract")
    print("  Ubuntu: sudo apt-get install tesseract-ocr")
    print("  Windows: Download from GitHub releases")
````

````python
# Function to convert PDF to images and extract text
def extract_text_from_pdf(pdf_path, dpi=300):
    """
    Convert PDF to images and extract text using OCR
    
    Args:
        pdf_path (str): Path to PDF file
        dpi (int): DPI for image conversion (higher = better quality, slower)
    
    Returns:
        str: Extracted text from all pages
    """
    print(f"📄 Converting PDF to images: {pdf_path}")
    
    # Convert PDF to images
    try:
        images = convert_from_path(pdf_path, dpi=dpi)
        print(f"✅ Converted to {len(images)} page(s)")
    except Exception as e:
        print(f"❌ Error converting PDF: {e}")
        return None
    
    # Extract text from each page
    all_text = []
    for i, image in enumerate(images):
        print(f"🔍 Processing page {i+1}/{len(images)}...")
        
        # Extract text using Tesseract
        try:
            text = pytesseract.image_to_string(image, lang='eng')
            all_text.append(f"--- PAGE {i+1} ---\n{text}\n")
            print(f"✅ Extracted {len(text)} characters from page {i+1}")
        except Exception as e:
            print(f"❌ Error extracting text from page {i+1}: {e}")
            all_text.append(f"--- PAGE {i+1} ---\n[ERROR: Could not extract text]\n")
    
    return "\n".join(all_text)

# Test the function
pdf_path = "/Users/andrewallkin/Downloads/Statement0885-2025-09-16.pdf"

if os.path.exists(pdf_path):
    extracted_text = extract_text_from_pdf(pdf_path)
    if extracted_text:
        with open("extracted_text.txt", "w") as f:
            f.write(extracted_text)
else:
    print(f"❌ PDF file not found: {pdf_path}")
    print("Please place your bank statement PDF in the same directory as this notebook")
````

````python
# Alternative: Extract text from a single image file
def extract_text_from_image(image_path):
    """
    Extract text from a single image file using OCR
    
    Args:
        image_path (str): Path to image file (PNG, JPG, etc.)
    
    Returns:
        str: Extracted text
    """
    print(f"🖼️ Processing image: {image_path}")
    
    try:
        # Open image
        image = Image.open(image_path)
        print(f"✅ Image loaded: {image.size[0]}x{image.size[1]} pixels")
        
        # Extract text
        text = pytesseract.image_to_string(image, lang='eng')
        print(f"✅ Extracted {len(text)} characters")
        
        return text
    except Exception as e:
        print(f"❌ Error processing image: {e}")
        return None

# Test with an image file
image_path = "sample_bank_statement.png"  # Replace with your image path

if os.path.exists(image_path):
    extracted_text = extract_text_from_image(image_path)
    if extracted_text:
        print("\n" + "="*50)
        print("EXTRACTED TEXT:")
        print("="*50)
        print(extracted_text)
else:
    print(f"❌ Image file not found: {image_path}")
    print("You can also test with any image file containing text")
````

````python
# Advanced OCR with different configurations
def extract_text_advanced(image_path, config='--psm 6'):
    """
    Extract text with custom Tesseract configuration
    
    Args:
        image_path (str): Path to image file
        config (str): Tesseract configuration options
    
    Returns:
        str: Extracted text
    """
    print(f"🔧 Advanced OCR with config: {config}")
    
    try:
        image = Image.open(image_path)
        
        # Different PSM modes for different text layouts:
        # PSM 3: Fully automatic page segmentation (default)
        # PSM 6: Single uniform block of text
        # PSM 8: Single word
        # PSM 13: Raw line. Treat the image as a single text line
        
        text = pytesseract.image_to_string(image, config=config, lang='eng')
        return text
    except Exception as e:
        print(f"❌ Error: {e}")
        return None

# Test different configurations
if os.path.exists(image_path):
    print("Testing different OCR configurations:")
    print("\n1. Default configuration:")
    text1 = extract_text_advanced(image_path, '--psm 3')
    print(text1[:200] + "..." if text1 and len(text1) > 200 else text1)
    
    print("\n2. Single text block:")
    text2 = extract_text_advanced(image_path, '--psm 6')
    print(text2[:200] + "..." if text2 and len(text2) > 200 else text2)
    
    print("\n3. Single line:")
    text3 = extract_text_advanced(image_path, '--psm 13')
    print(text3[:200] + "..." if text3 and len(text3) > 200 else text3)
````
